2. How do you interpret the coefficient for a dummy/one-hot-encoded variable? This is a trick question, and the trick involves how you handle the intercept of the model. There's further explanation at the end of this document, if needed.
    - you have to drop the first column if you have an intercept because that dropped one becomes the baseline or the starting point.
    - the coefficient for the other columns is just the difference from that dropped baseline. so if the intercept is the dropped row's value, the coefficient for a different row is just how much you add or subtract to get its specific value.

    - you can use categorical stuff as features (by one-hot encoding them), but the final outcome is always going to be a number, not a string or a category answer.
